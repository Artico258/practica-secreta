```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt

def convolucion_media(imagen, kernel):
    """Aplica convolución con el kernel dado"""
    imagen_c = cv2.filter2D(imagen, -1, kernel)
    return imagen_c
img = cv2.imread('joji.jpg')
if img is None:
    print("Error: No se pudo cargar la imagen 'mujer.jpg'")
    exit()
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
kernel_size = 35
kernel_h = np.ones((1, kernel_size), np.float32) / kernel_size
kernel_diag_izq = np.zeros((kernel_size, kernel_size), np.float32)
np.fill_diagonal(kernel_diag_izq, 1)
kernel_diag_izq = kernel_diag_izq / kernel_size
kernel_diag_der = np.zeros((kernel_size, kernel_size), np.float32)
np.fill_diagonal(np.fliplr(kernel_diag_der), 1)
kernel_diag_der = np.fliplr(kernel_diag_der)
kernel_diag_der = kernel_diag_der / kernel_size
img_horizontal = convolucion_media(img_rgb, kernel_h)
img_diag_izq = convolucion_media(img_rgb, kernel_diag_izq)
img_diag_der = convolucion_media(img_rgb, kernel_diag_der)
plt.figure(figsize=(18, 10))
plt.subplot(2, 2, 1)
plt.imshow(img_rgb)
plt.title('Imagen Original', fontsize=14, fontweight='bold')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.imshow(img_horizontal)
plt.title(f'Desenfoque Horizontal', fontsize=14, fontweight='bold')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.imshow(img_diag_izq)
plt.title(f'Desenfoque Diagonal Izquierda', fontsize=14, fontweight='bold')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.imshow(img_diag_der)
plt.title(f'Desenfoque Diagonal Derecha', fontsize=14, fontweight='bold')
plt.axis('off')

plt.tight_layout()
plt.show()
```
